```python
def plot_from_loc(outputs_loc):
    data_files = glob.glob(os.path.join(outputs_loc, 'data*.png'))
    data_files.sort()
    for data in data_files:
        correctly_labelled = data.replace("data", "correct")
        labels = data.replace("data", "labels")
        predictions = data.replace("data", "predictions")
        display(os.path.basename(data[:-4]))
        display(Image(data))
        display(Image(labels))
        display(Image(predictions))
        display(Image(correctly_labelled))
```
